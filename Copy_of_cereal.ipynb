```python
# Assuming cereal is a Pandas DataFrame
dc['mfr'] = dc['mfr'].astype('category')
dc['type'] = dc['type'].astype('category')

# Recode values
dc['mfr'].replace({
    'A': 'American Home Food Products',
    'G': 'General Mills',
    'K': 'Kelloggs',
    'N': 'Nabisco',
    'P': 'Post',
    'Q': 'Quaker Oats',
    'R': 'Ralston Purina'
}, inplace=True)

dc['type'].replace({
    'C': 'Cold',
    'H': 'Hot'
}, inplace=True)
```

```python
unique_mfr_levels = dc['mfr'].unique()
```

```python
unique_type_levels = dc['type'].unique()
print(unique_type_levels)
```

```python
# Assuming cereal is a Pandas DataFrame

# Select only numeric columns
dc_is_numeric = dc.select_dtypes(include=['number'])

# Calculate the correlation matrix
dc_correlation = dc_is_numeric.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(dc_correlation, annot=True, cmap="RdYlGn", fmt=".1f")
plt.title("Correlation Heatmap")
plt.xlabel("")
plt.ylabel("")
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming cereal is a Pandas DataFrame
dc_corr = dc_is_numeric.corr()['rating'].reset_index()
dc_corr = dc_corr[dc_corr['index'] != 'rating']

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(dc_corr['index'], dc_corr['rating'], color=['#3D2F38', '#715266', '#A3918C'])
plt.xlim(-1, 1)
plt.title("Correlation with Rating")
plt.xlabel("Correlation Value")
plt.ylabel("")
plt.gca().invert_yaxis()  # Flip y-axis to match coord_flip in ggplot
plt.show()
```

```python
# Assuming cereal is a Pandas DataFrame
dc_calcor = pd.melt(dc[['rating', 'calories', 'sugars']], id_vars='rating')

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dc_calcor, x='rating', y='value', hue='variable', palette=['#3D2F38', '#715266'])
sns.lmplot(data=dc_calcor, x='rating', y='value', hue='variable', ci=None, legend=False)
plt.title("Scatter Plot Correlation of Both Calories and Sugars with Ratings")
plt.xlabel("Rating")
plt.ylabel("")
plt.show()
```

```python


# Assuming cereal is a Pandas DataFrame
sugar_highest = dc[['name', 'sugars']].sort_values(by='sugars', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.barh(sugar_highest['name'], sugar_highest['sugars'], color=['#C1A5A9', '#694873'])
plt.axhline(y=sugar_highest['sugars'].mean(), color='white', linestyle='--', linewidth=1)
plt.text(0, sugar_highest['sugars'].mean(), round(sugar_highest['sugars'].mean(), 1), va='center')
plt.title("Top 10 Cereals with Most Sugar Content")
plt.xlabel("Sugar Content (grams)")
plt.ylabel("")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Assuming cereal is a Pandas DataFrame
mfr_sugar = dc.groupby('mfr')['sugars'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(mfr_sugar.index, mfr_sugar, color=['#5ADBFF', '#004BA8'])
plt.title("Average Sugar Content by Manufacturer")
plt.xlabel("Average Sugar Content (grams)")
plt.ylabel("Manufacturer")
plt.gca().invert_yaxis()
plt.show()
```
