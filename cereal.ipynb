```python
dc1c=dc1.corr()
sns.heatmap(dc1c, annot=True, cmap='coolwarm', linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
dc_rating=dc1.corrwith(dc1['rating'])
dc_rating.plot(kind='bar', color = 'blue')
plt.title('Rating Correlation with Colunms')
plt.xlabel('Colunms')
plt.ylabel('Correlation')
plt.show()
```
